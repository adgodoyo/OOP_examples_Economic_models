```python
class Disc_p :
    def __init__ (self, cmg , a , b): # Cada monopolista tendra una funcion de demanda diferente y un Cmg diferente, por lo cual estos son los atributos obligatorios de la clase
      assert cmg >= 0, "El CMg debe de ser positivo"
      assert a > 0, "La constante en la funci√≥n de demanda debe ser positva"
      assert b > 0, "Debe ingresar el valor absoluto de la pendiente y no puede ser totalmente el√°stica"

      self.corte = a
      self.pendiente = b
      self.costoMg = cmg

      self.Q_cmg = (a- cmg) / b #Atributos de clase (no se definen al crear el objeto)

    def Equilibrio_Mon(self) :
      q0 = (self.corte - self.costoMg) / (2 * self.pendiente)
      p0 = self.corte - (self.pendiente * (self.corte - self.costoMg) / (2 * self.pendiente))
      ex_cons_mon = (self.corte - p0) * (q0 / 2)
      ex_prod_mon = (p0 - self.costoMg) * q0
      pi_mon = ((self.Q_cmg - q0) * (p0 - self.costoMg)) / 2

      print(f"Excedente Consumidor: {ex_cons_mon} ")
      print(f"Excendete Productor: {ex_prod_mon}")
      print(f"Perdida Irrecuperable: {pi_mon}")
      return (q0, p0)

    def Discriminacion_grado_1 (self, q) :
      ex_cons_grd1 = 0
      ex_prod_grd1 = (self.corte - self.costoMg) * (self.Q_cmg)
      pi_grd1 = 0

      print(f"Excedente Consumidor: {ex_cons_grd1}")
      print(f"Excedente Productor: {ex_prod_grd1}")
      print(f"Perdida Irrecuperable: {pi_grd1}")
      if 0<= q <=  self.Q_cmg:
          return q, self.corte - (self.pendiente * q) #Me arroja el precio y la cantidad para cada individuo dependiendo la cantidad, discriminacion perfecta
      else:
          return 'La cantidad debe ser positiva y menor o igual a la cantidad m√°xima de producci√≥n'

    def Discriminacion_grado_2 (self, qq) :

      q1 = float(input(f"Ingrese la cantidad del primer segmento del mercado. Debe ser mayor a 0 y menor a la producci√≥n m√°xima de: {self.Q_cmg} "))
      while q1 <= 0 or q1 > self.Q_cmg:
        print("Cantidad no v√°lida. Por favor ingr√©sela nuevamente")
        q1 = float(input(f"Ingrese la cantidad del primer segmento del mercado. Debe ser mayor a 0 y menor a la producci√≥n m√°xima de: {self.Q_cmg} "))

      q2 = float(input(f"Ingrese la cantidad del segundo segmento del mercado. Debe ser mayor al primer segmento y menor a la producci√≥n m√°xima de: {self.Q_cmg} "))
      while q2 <= q1 or q2 > self.Q_cmg:
        print("Cantidad no v√°lida. Por favor ingr√©sela nuevamente")
        q2 = float(input(f"Ingrese la cantidad del segundo segmento del mercado. Debe ser mayor al primer segmento y menor a la producci√≥n m√°xima de: {self.Q_cmg} "))

      q3 = float(input(f"Ingrese la cantidad del tercer segmento del mercado. Debe ser mayor al segundo segmento y menor a la producci√≥n m√°xima de: {self.Q_cmg} "))
      while q3 <= q2 or q3 > self.Q_cmg:
        print("Cantidad no v√°lida. Por favor ingr√©sela nuevamente")
        q3 = float(input(f"Ingrese la cantidad del tercer segmento del mercado. Debe ser mayor al segundo segmento y menor a la producci√≥n m√°xima de: {self.Q_cmg} "))

      p1 = self.corte - (self.pendiente * q1)
      p2 = self.corte - (self.pendiente * q2)
      p3 = self.corte - (self.pendiente * q3)
      ex_cons_grd2 = ((self.corte - p1) * q1 / 2) + ((p1 -p2) * (q2 - q1) / 2) + ((p2 - p3) * (q3 - q2) / 2)
      ex_prod_grd2 = ((p1 - self.costoMg) * q1) + ((p2 - self.costoMg) * (q2 - q1)) + ((p3 - self.costoMg) * (q3 - q2))
      pi_grd2 = (self.Q_cmg - q3) * (p3 - self.costoMg) / 2

      print(f"Excedente Consumidor: {ex_cons_grd2}")
      print(f"Excedente Productor: {ex_prod_grd2}")
      print(f"Perdida Irrecuperable: {pi_grd2}")

      if 0 <= qq <= q1 :
        return qq, self.corte - (self.pendiente) * q1
      elif  q1 < qq <= q2 :
        return qq, self.corte - (self.pendiente) * q2
      elif q2 < qq <= q3 :
        return qq, self.corte - (self.pendiente) * q3
      elif qq > q3 :
        return "No hay suficiente producto para abastecer esa cantidad"
      else:
        return "La cantidad debe ser positiva"

    def Discriminacion_grado_3 (self, qqq) :
        grupo = input("Ingrese a que grupo del mercado pertenece su consumidor, grupo 1 o grupo 2 ")
        while grupo != "grupo 1" and grupo != "grupo 2" :
          print("Grupo no valido. Por favor ingreselo nuevamente")
          grupo = input("Ingrese a que grupo del mercado pertenece su consumidor, grupo 1 o grupo 2 ")

        if 0 <= qqq:
          if grupo == "grupo 1" :
            qq0 = (self.corte - self.costoMg) / (2 * self.pendiente)
            pp0 = self.corte - (self.pendiente * (self.corte - self.costoMg) / (2 * self.pendiente))
            ex_cons_grd3_1 = (self.corte - pp0) * (qq0 / 2)
            ex_prod_grd3_1 = (pp0 - self.costoMg) * qq0
            pi_grd3_1 = ((self.Q_cmg - qq0) * (pp0 - self.costoMg)) / 2

            print(f"Excedente Consumidor: {ex_cons_grd3_1} ")
            print(f"Excendete Productor: {ex_prod_grd3_1}")
            print(f"Perdida Irrecuperable: {pi_grd3_1}")

            if 0 <= qqq <= (self.corte - self.costoMg) / (2 * self.pendiente) :
              return qqq, (self.corte - (self.pendiente * (self.corte - self.costoMg) / (2 * self.pendiente)))*qqq
            else:
              return "No hay suficiente producto para abastecer esa cantidad "

          elif grupo == "grupo 2" :
            a1 = float(input("Ingrese el punto de corte de la funcion de demanda del grupo 2. Debe ser mayor a cero "))
            while a1 <= 0 :
              print("Punto de corte no valido. Por favor ingreselo nuevamente" )
              a1 = float(input("Ingrese el punto de corte de la funcion de demanda del grupo 2. Debe ser mayor a cero "))
            b1 = float(input("Ingrese el valor absoluto de la pendiente de la funcion de demanda del grupo 2. No puede ser una demanda totalmente elastica "))
            while b1 <= 0 :
              print("Pendiente no valida. Por favor ingresela nuevamente" )
              b1 = float(input("Ingrese el valor absoluto de la pendiente de la funcion de demanda del grupo 2. No puede ser una demanda totalmente elastica "))

            qq1 = (a1 - self.costoMg) / (2 * b1)
            pp1 = a1 - (b1 * qq1)
            Q_cmg1 = (a1 - self.costoMg) / b1
            ex_cons_grd3_2 = (a1 - pp1) * (qq1 / 2)
            ex_prod_grd3_2 = (pp1 - self.costoMg) * qq1
            pi_grd3_2 = ((Q_cmg1 - qq1) * (pp1 - self.costoMg)) / 2

            print(f"Excedente Consumidor: {ex_cons_grd3_2 } ")
            print(f"Excendete Productor: {ex_prod_grd3_2 }")
            print(f"Perdida Irrecuperable: {pi_grd3_2}")

            if 0 <= qqq <= (a1 - self.costoMg) / (2 * b1):
              return qqq, (a1 - (b1 * (a1 - self.costoMg) / (2 * b1)))*qqq
            else:
              return "No hay suficiente producto para abastecer esa cantidad "
        else:
          return "La cantidad no puede ser menor a cero"
```

# Grafica Equilibrio Monopolio


```python
import matplotlib.pyplot as plt
import numpy as np

class Disc_p:
    def __init__(self, cmg, a, b):
        assert cmg >= 0, "El CMg debe de ser positivo"
        assert a >= 0, "La constante en la funci√≥n de demanda debe ser positiva"
        assert b > 0, "Debe ingresar el valor absoluto de la pendiente y no puede ser totalmente el√°stica"

        self.corte = a
        self.pendiente = b
        self.costoMg = cmg

        self.Q_cmg = (a - cmg) / b  # Atributo de clase (cantidad en competencia perfecta)

    def Equilibrio_Mon(self):
        q0 = (self.corte - self.costoMg) / (2 * self.pendiente)
        p0 = self.corte - (self.pendiente * q0)
        ex_cons_mon = (self.corte - p0) * (q0 / 2)

        # Excedente del productor: √Årea entre el precio de equilibrio y el costo marginal
        ex_prod_mon = (p0 - self.costoMg) * q0

        # La p√©rdida irrecuperable es el √°rea del tri√°ngulo entre q0 y Q_cmg, debajo de la curva de demanda y sobre la l√≠nea CMg
        pi_mon = (self.Q_cmg - q0) * (self.corte - (self.pendiente * self.Q_cmg) - p0) / 2

        print(f"Excedente Consumidor: {ex_cons_mon}")
        print(f"Excedente Productor: {ex_prod_mon}")
        print(f"P√©rdida Irrecuperable: {pi_mon}")

        # Gr√°fica
        cantidades = np.linspace(0, self.corte / self.pendiente, 100)
        precios_demanda = self.corte - self.pendiente * cantidades

        plt.figure(figsize=(10, 6))

        # Curva de demanda
        plt.plot(cantidades, precios_demanda, label="Curva de Demanda", color="blue")

        # L√≠nea del CMG
        plt.axhline(y=self.costoMg, color='red', linestyle='--', label="CMg (Costo Marginal)")

        # √Årea del excedente del consumidor
        plt.fill_between(cantidades, precios_demanda, p0, where=(cantidades <= q0), color="green", alpha=0.3, label="Excedente del Consumidor")

        # √Årea del excedente del productor
        plt.fill_between(cantidades, self.costoMg, p0, where=(cantidades <= q0), color="orange", alpha=0.3, label="Excedente del Productor")

        # √Årea de la p√©rdida irrecuperable
        plt.fill_between(cantidades, self.corte - self.pendiente * cantidades, self.costoMg, where=(cantidades > q0) & (cantidades <= self.Q_cmg), color="purple", alpha=0.3, label="P√©rdida Irrecuperable")

        # Punto de equilibrio
        plt.scatter(q0, p0, color="black", label=f"Equilibrio Monopolio: Q={q0:.2f}, P={p0:.2f}")

        # Etiquetas y t√≠tulo
        plt.xlabel("Cantidad")
        plt.ylabel("Precio")
        plt.title("Excedentes y P√©rdida Irrecuperable en Monopolio")
        plt.legend()
        plt.grid(True)
        plt.show()

        return q0, p0

# Crear un objeto de la clase
ejemplo = Disc_p(2, 50, 4)

# Usar el m√©todo Equilibrio_Mon para generar la gr√°fica
ejemplo.Equilibrio_Mon()


```

```python
import matplotlib.pyplot as plt
import numpy as np

class Disc_p:
    def __init__(self, cmg, a, b):
        assert cmg >= 0, "El CMg debe de ser positivo"
        assert a >= 0, "La constante en la funci√≥n de demanda debe ser positiva"
        assert b > 0, "Debe ingresar el valor absoluto de la pendiente y no puede ser totalmente el√°stica"

        self.corte = a
        self.pendiente = b
        self.costoMg = cmg

        self.Q_cmg = (a - cmg) / b  # Atributos de clase

    def Discriminacion_grado_3(self, qqq):
        grupo = input("Ingrese a qu√© grupo del mercado pertenece su consumidor, grupo 1 o grupo 2: ")
        while grupo != "grupo 1" and grupo != "grupo 2":
            print("Grupo no v√°lido. Por favor ingr√©selo nuevamente.")
            grupo = input("Ingrese a qu√© grupo del mercado pertenece su consumidor, grupo 1 o grupo 2: ")

        if 0 <= qqq:
            if grupo == "grupo 1":
                qq0 = (self.corte - self.costoMg) / (2 * self.pendiente)
                pp0 = self.corte - (self.pendiente * qq0)
                ex_cons_grd3_1 = (self.corte - pp0) * (qq0 / 2)
                ex_prod_grd3_1 = (pp0 - self.costoMg) * qq0
                pi_grd3_1 = ((self.Q_cmg - qq0) * (pp0 - self.costoMg)) / 2

                print(f"Excedente Consumidor: {ex_cons_grd3_1}")
                print(f"Excedente Productor: {ex_prod_grd3_1}")
                print(f"P√©rdida Irrecuperable: {pi_grd3_1}")

                self.graficar_excedentes(qq0, pp0, ex_cons_grd3_1, ex_prod_grd3_1, pi_grd3_1, grupo, self.corte, self.pendiente)

                if 0 <= qqq <= qq0:
                    return qqq, pp0 * qqq
                else:
                    return "No hay suficiente producto para abastecer esa cantidad."

            elif grupo == "grupo 2":
                a1 = float(input("Ingrese el punto de corte de la funci√≥n de demanda del grupo 2. Debe ser mayor a cero: "))
                while a1 <= 0:
                    print("Punto de corte no v√°lido. Por favor ingr√©selo nuevamente.")
                    a1 = float(input("Ingrese el punto de corte de la funci√≥n de demanda del grupo 2. Debe ser mayor a cero: "))
                b1 = float(input("Ingrese el valor absoluto de la pendiente de la funci√≥n de demanda del grupo 2. No puede ser una demanda totalmente el√°stica: "))
                while b1 <= 0:
                    print("Pendiente no v√°lida. Por favor ingr√©sela nuevamente.")
                    b1 = float(input("Ingrese el valor absoluto de la pendiente de la funci√≥n de demanda del grupo 2. No puede ser una demanda totalmente el√°stica."))

                qq1 = (a1 - self.costoMg) / (2 * b1)
                pp1 = a1 - (b1 * qq1)
                Q_cmg1 = (a1 - self.costoMg) / b1
                ex_cons_grd3_2 = (a1 - pp1) * (qq1 / 2)
                ex_prod_grd3_2 = (pp1 - self.costoMg) * qq1
                pi_grd3_2 = ((Q_cmg1 - qq1) * (pp1 - self.costoMg)) / 2

                print(f"Excedente Consumidor: {ex_cons_grd3_2}")
                print(f"Excedente Productor: {ex_prod_grd3_2}")
                print(f"P√©rdida Irrecuperable: {pi_grd3_2}")

                self.graficar_excedentes(qq1, pp1, ex_cons_grd3_2, ex_prod_grd3_2, pi_grd3_2, grupo, a1, b1)

                if 0 <= qqq <= qq1:
                    return qqq, pp1 * qqq
                else:
                    return "No hay suficiente producto para abastecer esa cantidad."
        else:
            return "La cantidad no puede ser menor a cero."

    def graficar_excedentes(self, cantidad, precio, ex_cons, ex_prod, pi, grupo, corte, pendiente):
        cantidades = np.linspace(0, self.Q_cmg if grupo == "grupo 1" else (corte - self.costoMg) / pendiente, 100)
        precios_demanda = corte - pendiente * cantidades

        plt.figure(figsize=(10, 6))

        # Curva de demanda
        plt.plot(cantidades, precios_demanda, label=f"Curva de Demanda ({grupo})", color="blue")

        # L√≠nea del CMG
        plt.axhline(y=self.costoMg, color='red', linestyle='--', label="CMg (Costo Marginal)")

        # Excedente del consumidor
        plt.fill_between([0, cantidad], [corte, precio], precio, color="green", alpha=0.5, label="Excedente del Consumidor")

        # Excedente del productor
        plt.fill_between([0, cantidad], precio, self.costoMg, color="orange", alpha=0.5, label="Excedente del Productor")

        # P√©rdida irrecuperable
        plt.fill_between([cantidad, self.Q_cmg if grupo == "grupo 1" else (corte - self.costoMg) / pendiente], [precio, self.corte - self.pendiente * self.Q_cmg], self.costoMg, color="purple", alpha=0.5, label="P√©rdida Irrecuperable")

        # Puntos de equilibrio
        plt.scatter(cantidad, precio, color="black", label=f"Equilibrio: Q={cantidad:.2f}, P={precio:.2f}")

        # Etiquetas y t√≠tulo
        plt.xlabel("Cantidad")
        plt.ylabel("Precio")
        plt.title(f"Discriminaci√≥n de Precios de Tercer Grado ({grupo})")
        plt.legend()
        plt.grid(True)
        plt.show()

# Crear un objeto de la clase
ejemplo = Disc_p(2, 50, 4)

# Usar el m√©todo Discriminacion_grado_3 para calcular y graficar
ejemplo.Discriminacion_grado_3(4)
```
